```python
lb = LabelEncoder()
```

```python
df["Make_encode"] = lb.fit_transform(df["Make"])
df["Model_Transmission_Type"] = lb.fit_transform(df["Transmission Type"])
```

```python
df["Model_Driven_Wheels"] = lb.fit_transform(df["Driven_Wheels"])
```

```python
df["Vehicle Style_encode"] = lb.fit_transform(df["Vehicle Style"])
```

```python
df
```

```python
# Creating ML models
df.columns
```

```python
X = df[['Make_encode', 'Model_encode',
       'Model_Engine Fuel Type', 'Model_Transmission_Type',
       'Model_Driven_Wheels', 'Vehicle Size_encode', 'Vehicle Style_encode',
       'Year','Engine HP','Engine Cylinders','highway MPG',
       'city mpg','Popularity']]
y = df['MSRP']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
```

```python
lr.fit(X_train,y_train)
```

```python
rf.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
rf.score(X_test,y_test)
```

```python
rf.score(X_train,y_train)
```
